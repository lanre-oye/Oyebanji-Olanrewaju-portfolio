```python
df = df.dropna(axis='columns', how='all')
df ## df shows you the whole dataframe
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df= df.drop(columns=['Last Update','Entry Date','Admin Units','Start Day','End Day','Magnitude','Origin','External IDs'])
```

```python
df.info()
```

```python
column_to_drop =[17,16,25,26,0,1,2,3,4,5,14,15,11] ## removing columns thorugh index numbers due to the pecularity of their names and other inrelevant error
df = df.drop(df.columns[column_to_drop],axis=1)
```

```python
df.info()
```

```python
Column_to_fill =['Start Month','End Month','Total Affected','No. Affected','Total Deaths']
for column_name in Column_to_fill:
  df.loc[df[column_name].isna(),column_name] = 0
  df.head()
```

```python
df.head()
```

```python
Column_2_fill =['Associated Types']
for column_name in Column_2_fill:
  df.loc[df[column_name].isna(),column_name] = 'Not_Specified'
  df.head()
```

```python
df.head()
```

```python
df.info()
```

# How to change data type

```python
Type_dictionary = {
    'Start Year': 'category',
    'End Year': 'category',
    'Start Month': 'category',
    'End Month': 'category',
    'Total Affected': 'int64',
    'No. Affected': 'int64',
    'Total Deaths': 'int64'
}
for key,value in Type_dictionary.items():
  df[key] = df[key].astype(value)
  df.info()
```
